```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('LABORDATEN.csv')
data1.head()
```

```python
# Liste der relevanten Spalten für die Nährstoffanalyse
relevante_spalten = [
    'PointID',
    'pH_H2O',
    'EC_H2O',
    'pH_CaCl2',
    'TC',
    'TOC',
    'TIC',
    'TN',
    'TS',
    'Mn_Ox',
    'Fe_Ox',
    'Al_Ox',
    'Mn_Di',
    'Fe_Di',
    'Al_Di',
    # Optional:
    'Wassergehalt',
    'TRD_FB',
    'FBV'
]

# Nur relevante Spalten behalten (überschneidet mit vorhandenen Spalten)
vorhandene_spalten = [spalte for spalte in relevante_spalten if spalte in data1.columns]
data1_necessary = data1[vorhandene_spalten]

# Ergebnis speichern oder weiterverarbeiten
data1_necessary.to_csv("bodendaten_naehrstoffe.csv", index=False)

print("Datensatz erfolgreich reduziert auf Nährstoff-relevante Parameter.")
```

```python
data2 =pd.read_csv('HORIZONTDATEN.csv')
data2.head()
```

```python
# Liste der relevanten Spalten für die Nährstoffanalyse
relevante_spalten = [
    'PointID',
    'Karbonat',
    'Humusgehaltsklasse',
    'Durchwurzelung',

]

# Nur relevante Spalten behalten (überschneidet mit vorhandenen Spalten)
vorhandene_spalten = [spalte for spalte in relevante_spalten if spalte in data2.columns]
data2_necessary = data2[vorhandene_spalten]

# Ergebnis speichern oder weiterverarbeiten
data2_necessary.to_csv("HORIZONTDATEN_red.csv", index=False)

print("Datensatz erfolgreich reduziert auf Nährstoff-relevante Parameter.")
```

```python
data3 = pd.read_csv('STANDORT.csv')
data3.head()
```

```python
# Liste der relevanten Spalten für die Nährstoffanalyse
relevante_spalten = [
    'PointID',
    'Bundesland',
    'Probenahme_Monat',
    'Probenahme_Jahr',
   
]

# Nur relevante Spalten behalten (überschneidet mit vorhandenen Spalten)
vorhandene_spalten = [spalte for spalte in relevante_spalten if spalte in data3.columns]
data3_necessary = data3[vorhandene_spalten]

# Ergebnis speichern oder weiterverarbeiten
data3_necessary.to_csv("STANDORT_red.csv", index=False)

print("Datensatz erfolgreich reduziert auf zeitlich-relevante Parameter.")
```

```python
df1 = pd.read_csv('LABORDATEN_red.csv')
df2 = pd.read_csv('HORIZONTDATEN_red.csv')
df3 = pd.read_csv('STANDORT_red.csv')
```

```python
merged = pd.merge(df1, df2, on="PointID", how="left")  # Ortsinfo dazukleben
merged = pd.merge(merged, df3, on="PointID", how="left")  # Zeitinfo dazukleben
```
