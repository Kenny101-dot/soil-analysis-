```python
import pandas as pd
import numpy as np
```

```python
data1 = pd.read_csv('LABORDATEN.csv')
data1.head()
```

```python
# Liste der relevanten Spalten für die Nährstoffanalyse
relevante_spalten = [
    'PointID',
    'pH_H2O',
    'EC_H2O',
    'pH_CaCl2',
    'TC',
    'TOC',
    'TIC',
    'TN',
    'TS',
    'Mn_Ox',
    'Fe_Ox',
    'Al_Ox',
    'Mn_Di',
    'Fe_Di',
    'Al_Di',
    # Optional:
    'Wassergehalt',
    'TRD_FB',
    'FBV'
]

# Nur relevante Spalten behalten (überschneidet mit vorhandenen Spalten)
vorhandene_spalten = [spalte for spalte in relevante_spalten if spalte in data1.columns]
data1_necessary = data1[vorhandene_spalten]

# Ergebnis speichern oder weiterverarbeiten
data1_necessary.to_csv("bodendaten_naehrstoffe.csv", index=False)

print("Datensatz erfolgreich reduziert auf Nährstoff-relevante Parameter.")
```

```python
df1 = pd.read_csv('LABORDATEN_red.csv')
df2 = pd.read_csv('HORIZONTDATEN_red2.csv')
df3 = pd.read_csv('STANDORT_red.csv')
```

```python
merged = pd.merge(df1, df2, on="PointID", how="left")  # Ortsinfo dazukleben
merged = pd.merge(merged, df3, on="PointID", how="left")  # Zeitinfo dazukleben
```

```python
merged.to_csv("combined_data.csv", index=False)
print("The merged dataset has been saved as 'combined_data.csv'.")
```

```python
df1_agg = df1.groupby("PointID").mean(numeric_only=True).reset_index()
```

```python
# Schritt 1: df2 bereinigen (1 Zeile pro PointID)
df2_clean = df2.drop_duplicates(subset='PointID')

# Schritt 2: Zeitdaten auch bereinigen (falls nicht eindeutig)
df3_clean = df3.drop_duplicates(subset='PointID')

# Schritt 3: Von df2 aus linken
merged = df2_clean \
    .merge(df1_agg, on="PointID", how="left") \
    .merge(df3_clean, on="PointID", how="left")
```

```python
# Remove 'C' from Karbonat, 'h' from Humusgehaltsklasse, and 'W' from Durchwurzelung
df2['Karbonat'] = df2['Karbonat'].str.replace('C', '', regex=False)
df2['Humusgehaltsklasse'] = df2['Humusgehaltsklasse'].str.replace('h', '', regex=False)
df2['Durchwurzelung'] = df2['Durchwurzelung'].str.replace('W', '', regex=False)

# Replace NaN values with 0
df2 = df2.fillna(0)
```

```python
df2['Karbonat'] = pd.to_numeric(df2['Karbonat'], errors='coerce')
df2['Humusgehaltsklasse'] = pd.to_numeric(df2['Humusgehaltsklasse'], errors='coerce')
df2['Durchwurzelung'] = pd.to_numeric(df2['Durchwurzelung'], errors='coerce')
```
